```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# Load the dataset
data = pd.read_csv('/content/energydata_complete.csv')

# Remove specified columns: "date" and "lights"
data = data.drop(columns=['date', 'lights'])

# Define the target variable ("Appliances")
y = data['Appliances']

# Define the features (independent variables) by excluding the target variable
X = data.drop(columns=['Appliances'])

# Split the data into training and test sets (70-30 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalize the data using MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Fit a multiple linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the training set
y_train_pred = model.predict(X_train)

# Calculate MAE for the training set
mae = mean_absolute_error(y_train, y_train_pred)

# Print the MAE
print("Mean Absolute Error (MAE) for the training set: {:.3f}".format(mae))
```

```python
# Calculate RMSE for the training set
rmse = mean_squared_error(y_train, y_train_pred, squared=False)
rmse
```

```python
# Make predictions on the test set
y_test_pred = model.predict(X_test)

# Calculate MAE for the test set
mae = mean_absolute_error(y_test, y_test_pred)

# Print the MAE with three decimal places
print("Mean Absolute Error (MAE) for the test set: {:.3f}".format(mae))
```
